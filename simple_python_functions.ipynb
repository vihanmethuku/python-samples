```python
import calendar

def display_monthly_calendar(year, month):
  """
  Displays the calendar for a given month and year.

  Args:
    year (int): The year (e.g., 2024).
    month (int): The month (1 for January, 12 for December).
  """
  if not (1 <= month <= 12):
    print("Error: Month must be between 1 and 12.")
    return

  print(f"\n--- Calendar for {calendar.month_name[month]} {year} ---")
  print(calendar.month(year, month))
  print("------------------------------------\n")

# --- Example Usage ---

# Example 1: Display calendar for May 2024
display_monthly_calendar(2024, 5)

# Example 2: Display calendar for December 2025
display_monthly_calendar(2025, 12)

# Example 3: Display calendar for February 2023 (to see a shorter month)
display_monthly_calendar(2023, 2)

# Example 4: Invalid month input
display_monthly_calendar(2024, 13)
```
